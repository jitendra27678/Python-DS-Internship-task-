#### Problem Statement 1:Smart Parking Lot Management System


```python
def parking_system(capacity, logs):
    current = 0
    peak = 0
    
    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1
        
        if current > peak:
            peak = current
    
    print("Currently Parked Vehicles:", current)
    
    if current > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")
    
    


parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

#### Problem Statement 2:Online Food Delivery Time Estimator


```python
def delivery_time(distance, traffic, weather):
    time = distance * 5   
    
    if traffic == "High":
        time += 10
    elif traffic == "Medium":
        time += 5
    
    if weather == "Rainy":
        time += 5
    elif weather == "Storm":
        time += 10
    
    print("Estimated Delivery Time:", time, "minutes")


delivery_time(8, "High", "Rainy")
```

#### Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def seat_analysis(total_seats, booked):
    booked_count = len(booked)
    
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")   
    
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

seat_analysis(200, [1]*150)
```

#### Problem Statement 4:Cloud Server Load Classification System


```python
def server_load(cpu_readings):
    total = 0
    
    for c in cpu_readings:
        total += c
    
    avg = total / len(cpu_readings)
    
    print(f"Average CPU Load: {int(avg)}%")
    
    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")



server_load([45, 60, 70, 85, 90])
```

#### Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def resource_monitor(resources):
    overused = []
    
    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)
    
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resources)
```

#### Problem Statement 6: Online Event Registration Capacity Controller


```python
def event_registration(capacity, registrations):
    
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


event_registration(100, 105)
```
