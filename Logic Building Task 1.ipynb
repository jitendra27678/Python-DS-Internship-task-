2. Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]

pass_count=0
fail_count=0

for mark in marks:
    if mark >=50:
        pass_count += 1
    else:
        fail_count += 1
print("Total Passed Students:", pass_count)
print("Total Failed Students:", fail_count)
```

3. Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    clean_name = name.strip().lower()
    cleaned_names.append(clean_name)

print("Cleaned Names:", cleaned_names)
```

4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length =len(msg)
    print(f"message: {msg}, Length : {length}")

    if length  > 10:
        print("This is a long message")
```
