#### Problem 1: Mobile Recharge Validation System


```python
valid_plans = [199, 299, 399, 599]

def recharge_validation(valid_plans):
    while True:
        amount = int(input("Enter recharge amount: "))
        
        if amount >= 50 and amount in valid_plans:
            print("Recharge Successful ")
            break
        else:
            print("Invalid recharge Try again")

recharge_validation(valid_plans)
```

#### Problem 2: Inventory Reorder Alert System


```python
products = {"Rice": 10, "Wheat": 20, "Sugar": 8}

def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")

check_inventory(products)
```

#### Problem 3: Student Result Processing System


```python
marks = [60, 70, 40]
def student_result(marks):
    total = 0
    
    for m in marks:
        total += m
    
    avg = total / len(marks)
    
    if avg >= 50:
        return f"Average: {avg}  Pass"
    else:
        return f"Average: {avg}  Fail"
student_result(marks)
```

#### Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak:
        fare *= 1.25
    return fare

while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ").lower()
    
    is_peak = True if peak == "yes" else False
    fare = calculate_fare(km, is_peak)
    
    print(f"Total Fare: ₹{fare:.2f}")
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

#### Problem 5: Employee Attendance Eligibility Checker


```python
attendance = ["P", "P", "A", "P", "P"]
def check_attendance(attendance):
    present = 0
    
    for day in attendance:
        if day == "P":
            present += 1
    
    percentage = (present / len(attendance)) * 100
    
    if percentage >= 75:
        return f"{percentage}% → Eligible"
    else:
        return f"{percentage}% → Not Eligible"

attendance = ["P", "P", "A", "P", "P"]
check_attendance(attendance)
```

#### Problem 6: Password Strength Checker


```python
password= input("Enter password:")
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    if len(password) < 8:
        return "Weak Password ❌"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password "
check_password(password)
```
